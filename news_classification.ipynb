```python
'''引入 trainingset'''
import glob
import os
import pandas as pd
fnames = glob.glob("chinese_news_trans/*/*.txt")
datas = {
    "content":[],
    "target": []
}
for fn in fnames:
    with open(fn, "r", encoding="utf-8") as f:
        content = f.read()
        datas["content"].append(content)
        dn = os.path.split(fn)[0]
        dn = os.path.split(dn)[-1]
        datas["target"].append(dn)
train_df = pd.DataFrame(datas)
train_df
```

```python
'''引入 testset'''
fnames = glob.glob("chinese_news_test/*/*.txt")
datas = {
    "content":[],
    "target": []
}
for fn in fnames:
    with open(fn, "r", encoding="utf-8") as f:
        content = f.read()
        datas["content"].append(content)
        dn = os.path.split(fn)[0]
        dn = os.path.split(dn)[-1]
        datas["target"].append(dn)
test_df = pd.DataFrame(datas)
test_df
```

```python
'''import jieba 分詞 & 官方BIG5辭典庫'''
import jieba
from urllib.request import urlretrieve
# with open("news.txt", "r", encoding="utf-8") as f:
#     news = f.read()

url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
urlretrieve(url, "dict.txt.big")

jieba.set_dictionary("dict.txt.big")
# jieba.load_userdict("health_dict.txt")

# " ".join(jieba.cut(news))
```

```python
import jieba.analyse
train_df["content"]
content = ""
for txt in train_df["content"]:
    content += txt
jieba.analyse.extract_tags(content)
```

```python
# 答案: 處理成數字

u = train_df["target"].unique()
trans = {name: i for i, name in enumerate(u)} # {'交通': 0, '政治': 1, '教育': 2, '環境': 3, '經濟': 4, '藝術': 5, '計算機': 6, '軍事': 7, '醫藥': 8, '體育': 9}
reverse_trans = {i:name  for name, i in trans.items()} # diction.items() return tuple(k,v)
print(trans)
print(reverse_trans)
y_train = train_df["target"].replace(trans)
y_test = test_df["target"].replace(trans)
# print(y_train)
print(y_test)
```

```python
'''標點符號處理'''
def newscut(s):
    s = " ".join(jieba.cut(s))
    s.replace("\r", "").replace("\n", "")
    return s

# pandas 表格: DataFrame 一欄或者一列: Series
# Series.apply(功能名字)
x_train = train_df["content"].apply(newscut)
x_test = test_df["content"].apply(newscut)
x_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x_train_vec = vec.fit_transform(x_train)
x_test_vec = vec.transform(x_test)
x_train_vec # sparse matrix 存非零的值
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_vec)
print("正確率:", accuracy_score(pre, y_test))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pre)
c = ["{}(預測)".format(name) for name in u]
i = ["{}(原本)".format(name) for name in u]
pd.DataFrame(mat, columns=c, index=i)
```
