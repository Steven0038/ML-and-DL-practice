### 維度災難
> 多欄位變量 EX:NLP 分詞, 使用傳統演算法EX:決策樹  會產生維度災難

> ex: 100 petal length, 100 pedal width, 二維度空間會是 100*100=10000, 如果是三維就是 100*100*100=1000000

### 條件機率

> P(A|B): under B, A 發生機率

> P(A,B): A,B 同時發生的機率 = P(A) * P(B|A) = P(B) * P(A|B); P(B|A) = P(B) * P(A|B) / P(A)

> 一般我們做機率連算連乘(Ex: 骰子連續骰出哪一面的機率), 是指獨立事件的機率; 而如果事件間會互相影響(非獨立事件), 則需要用條件事件去算

> 李白寫詩寫出 我->喜歡->你; 不可能是獨立事件~~~~~ 所以需要使用條件機率, 但是如果事情簡化成單純猜單詞出現的機率(錯誤假設)
> 就可以使用單純貝氏, 單純在不同字匯在詩詞資料中出現的機率相乘, 來比較不同詩人寫出(骰出) 我->喜歡->你 的機率

#### 比較 P(李白|我->喜->你) P(杜甫|我->喜->你) P(白居易|我->喜->你)
>P(李白|我->喜->你) = P(李白) * P(我->喜->你/李白) / P(我->喜->你)  => 分母一樣可以只比較 P(我->喜->你/李白) or P(我->喜->你/杜甫) P(我->喜->你/白居易)

> 事實上用單純貝氏,　也有部分與詞向量加權相似的特性，比如說, P(我|李白) P(我|杜甫), 因為我是常用詞, 事實上這二個機率非常相似而在比較時會互相約掉

> 特徵工程 => 通常在選擇演算法前會先決定特徵重要性, 但在事前刪除要非常小心, 最好是讓演算法自己決定重要性, 再事後刪除不重要的特徵

> 在這次使用 NB, 不需要自己選擇特徵, NB 會自己決定特徵(出現機率)

#### https://scikit-learn.org/stable/modules/naive_bayes.html  不同的 naive bayes 演算法

> multinomial NB => 骰子骰幾次就是做幾次方; 

> Gaussian NB => 條件機率獨立事件 用有限的資料做出每個點的機率(高斯/常態分布), 主要用在數值是連續型的題目(Ex: iris)

> Bernoulli NB => P, 1-P 特殊形態的骰子 (不常用), 主要用在 2 值型題目  

> 做貝氏機率的時候, 需要做 smoothing, 讓特徵次數趨近於0 或趨近於1, 但不要是0 or 1, 這邊是做 Laplace smoothing, 避掉在計算機率時乘零造成結果為零的問題

```python
import pandas as pd
train_df = pd.read_csv("./poem/poem_train.csv", encoding="utf-8")
train_df
```

```python
test_df = pd.read_csv("./poem/poem_test.csv", encoding="utf-8")
test_df
```

```python
# 答案: 處理成數字
# replace()
# trans = {
#     "李白": 0,
#     "杜甫": 1,
#     "白居易": 2
# }

u = train_df["作者"].unique()
trans = {name: i for i, name in enumerate(u)} # {'李白': 0, '杜甫': 1, '白居易': 2}
reverse_trans = {i:name  for name, i in trans.items()} # diction.items() return tuple(k,v)
print(trans)
print(reverse_trans)
y_train = train_df["作者"].replace(trans)
y_test = test_df["作者"].replace(trans)
# print(y_train)
print(y_test)
```

```python
# 如果詩詞分類, 標點符號和換行, 應該去掉, 因為不影響結果
import jieba
s = "平林漠漠煙如織,平林漠漠煙如織，寒山一帶傷心碧。"

s = " ".join(jieba.cut(s))
# s.replace("\r", "").replace("\n", "")
```

```python
def poemcut(s):
    s = " ".join(jieba.cut(s))
    s.replace("\r", "").replace("\n", "")
    return s

# pandas 表格: DataFrame 一欄或者一列: Series
# Series.apply(功能名字)
x_train = train_df["內容"].apply(poemcut)
x_test = test_df["內容"].apply(poemcut)
x_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x_train_vec = vec.fit_transform(x_train)
x_test_vec = vec.transform(x_test)
x_train_vec # sparse matrix 存非零的值
```

```python
# print(x_train_vec) #(第幾首詩, 第幾個詞, 出現幾次)
# vec.vocabulary_ # 第幾首詩的代號
# vec.vocabulary_["傷心"] # 檢查有沒有在裡面
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_vec)
print("正確率:", accuracy_score(pre, y_test))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pre)
c = ["{}(預測)".format(name) for name in u]
i = ["{}(原本)".format(name) for name in u]
pd.DataFrame(mat, columns=c, index=i)
```
